```python
# Calculate the earliest and latest dates in the dataset
earliest_date = all_data['Order Date'].min()
latest_date = all_data['Order Date'].max()

# Define a 3-month period as the frequency
period_days = 365

# Function to check if a customer is a repeat customer within each 3-month period
def is_repeat_customer(group):
    # Calculate the difference in days between the earliest and latest dates
    days_difference = (group['Order Date'].max() - group['Order Date'].min()).days
    return days_difference <= period_days

# Group by 'CustomerKey' and apply the 'is_repeat_customer' function
repeat_customers = all_data.groupby('CustomerKey').filter(is_repeat_customer)
```

```python
repeat_customers
```

```python
repeat_customers['CustomerKey'].nunique()
```

```python
max_order_date = all_data.groupby('CustomerKey')['Order Date'].max().reset_index()
second_last_order_date = all_data.groupby('CustomerKey')['Order Date'].apply(lambda x: x.nlargest(2).iloc[-1]).reset_index()

# Merge the two DataFrames to get the maximum and second-to-last order date for each customer
merged_dates = pd.merge(max_order_date, second_last_order_date, on='CustomerKey', suffixes=('_max', '_second_last'))

# Calculate the difference in days between the maximum and second-to-last order date
merged_dates['days_difference'] = (merged_dates['Order Date_max'] - merged_dates['Order Date_second_last']).dt.days

# Filter customers who made a purchase within a 3-month period (90 days)
repeat_customers_1 = merged_dates[merged_dates['days_difference'] <= 365]

# Display the DataFrame with repeat customers
print(repeat_customers_1)
```

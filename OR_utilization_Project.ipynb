```python
df['Actual_surgery_duration']= df['End Time']-df['Start Time']
```

```python
df
```

```python
df['Booked Time (min)'].max()
```

```python
def timedelta_to_minutes(td):
    return td.components.hours * 60 + td.components.minutes + td.components.seconds / 60
```

```python
df['actual_d']= df['Actual_surgery_duration'].apply(timedelta_to_minutes)
```

```python
df.info()
```

```python
df.drop(columns={'Actual_surgery_duration'}, inplace=True)
```

```python
df
```

```python
df['difference']=df['Booked Time (min)']- df['actual_d']
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(x='Service', y='difference', data=df)
```

```python
df[df['Service']=='OBGYN']
```

```python
df.pivot_table(values='difference', index='Service', aggfunc=['mean','max','min'])
```

```python
df.describe()
```

Looking at pivot table above we can see 'Plastic' and 'OBGYN' have high delays i.e. overbooked OT whereas 'Vascular' and 'Opthalmology' have the least delays
